```python
def func(x):
  return x**3 - x**2 + 2

def bisection(a,b):
  if func(a) * func(b) >=0:
    print("wrong a and b")
    return
  c=a
  while(b-a) >= 0.01:
    c= (a+b)/2
    if func(c)==0.0:
      break
    if func(c) * func(a) < 0:
      b=c
    else:
      a=c
  print("Root:",format(c,".4f"))

a=float(input("Enter a:"))
b=float(input("Enter b:"))
bisection(a,b)
```
