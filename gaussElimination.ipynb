```python
def gauss_elimination():
    n = int(input("Enter number of equations: "))
    mat = []
    for i in range(n):
        row = list(map(float, input(f"Enter row {i+1} (with b): ").split()))
        mat.append(row)

    for i in range(n):
        for j in range(i+1, n):
            ratio = mat[j][i] / mat[i][i]
            for k in range(n+1):
                mat[j][k] -= ratio * mat[i][k]

    x = [0.0] * n
    for i in range(n-1, -1, -1):
        x[i] = mat[i][n]
        for j in range(i+1, n):
            x[i] -= mat[i][j] * x[j]
        x[i] /= mat[i][i]

    for i in range(n):
        print(f"x{i+1} =", format(x[i], ".4f"))

gauss_elimination()
```
